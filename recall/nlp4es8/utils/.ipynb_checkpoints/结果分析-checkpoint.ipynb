```python
import requests
import json
def get_match_text(c_question):
    ss=requests.post('http://localhost:9004/FindAnswer?question={}&top_n=10&original=weibao'.format(c_question))
    json_result=eval(json.loads(ss.content))
    results=[]
    for result in json_result:
        question=result['question']
        primary_id=result['primaryQuestionId']
        score =result['score']
        results.append((c_question,question,primary_id,score))
    results.sort(key=lambda x:x[3])
    results.reverse()
    return results
```

```python
import jieba
def load_stopwords():
    with open('stopwords.txt',"r",encoding="utf-8") as f:
        stopwords = {}
        for line in f:
            stopwords[line.strip()] = 0
    return stopwords
stopwords=load_stopwords()
class Cal_Similarity(object):

    # 分词，剔除无意义词语
    def seg_sentence(self, sentences):
        sentence_seged = jieba.lcut(sentences)

        # load stopwords
        outstr = []
        for word in sentence_seged:
            if word not in stopwords:
                if word != '\t':
                    outstr.append(word)

        return outstr

    def LCS(self, A, B):
        n = len(A)
        m = len(B)

        # 在字符串数组A、B之前插入字符0，目的是使后面下标统一
        A.insert(0, '0')
        B.insert(0, '0')

        # 二维表L存放公共子序列的长度
        L = [([0] * (m + 1)) for i in range(n + 1)]

        # 二维表C存放公共子序列的长度步进
        C = [([0] * (m + 1)) for i in range(n + 1)]

        for x in range(0, n + 1):
            for y in range(0, m + 1):
                if (x == 0 or y == 0):
                    L[x][y] = 0
                elif A[x] == B[y]:
                    L[x][y] = (L[x - 1][y - 1] + 1)
                    C[x][y] = 0
                elif L[x - 1][y] >= L[x][y - 1]:
                    L[x][y] = L[x - 1][y]
                    C[x][y] = 1
                else:
                    L[x][y] = L[x][y - 1]
                    C[x][y] = -1
        return L[n][m], C, n, m

    def printLCS(self, C, A, x, y):
        if (x == 0 or y == 0):
            return 0
        if C[x][y] == 0:
            self.printLCS(C, A, x - 1, y - 1)
        elif C[x][y] == 1:
            self.printLCS(C, A, x - 1, y)
        else:
            self.printLCS(C, A, x, y - 1)

    def str_to_list(self, str):
        result = []
        sub_str = ""
        for i in range(len(str)):
            coding = ord(str[i:i + 1])
            if 48 <= coding <= 57 or 65 <= coding <= 57 or 97 <= coding <= 122:
                sub_str += str[i:i + 1]
            else:
                if len(sub_str) == 0:
                    result.append(str[i:i + 1])
                else:
                    result.append(sub_str)
                    sub_str = ""
        return result

    def seg_score(self, A, B):
        A_list = self.seg_sentence(A)
        B_list = self.seg_sentence(B)

        A_and_B = list(set(A_list).intersection(set(B_list)))
        A_or_B = list(set(A_list).union(B_list))

        jarrcard_score = len(A_and_B) * 1.0 / len(A_or_B)

        return jarrcard_score

    def cal_similar(self, A, B):
        # 计算LCS分值
        A_list = self.str_to_list(A)
        B_list = self.str_to_list(B)

        length, C, x, y = self.LCS(A_list, B_list)

        A_len = len(A_list)
        B_len = len(B_list)

        min_len = 0
        max_len = 999
        if A_len > B_len:
            min_len = B_len
            max_len = A_len
        else:
            min_len = A_len
            max_len = B_len

        lcs_score = round(length / min_len, 5)

        # 计算jarcarrd分值
        jac_score = round(self.seg_score(A, B), 5)

        # 分值加权,取最大值
        if lcs_score > jac_score:
            return lcs_score
        else:
            return jac_score

    def cal_cos_sim(self, base_text_vec, compaired_text_vec):

        norm_1 = np.linalg.norm(base_text_vec)
        norm_2 = np.linalg.norm(compaired_text_vec)

        dot_product = np.dot(base_text_vec, compaired_text_vec)

        similarity = round(0.5 + 0.5 * (dot_product / (norm_1 * norm_2)), 5)

        return similarity
```

```python
from tqdm.notebook import tqdm
def single_pass(corpus, theta):
    """
    2020-6-5 将single_pass聚类结果平均句向量，然后计算置信度，获取置信度
    :param corpus_vec:
    :param corpus:
    :param theta:
    :return:
    """
    cluster_text = {}
    num_topic = 0

    # 若数据量太少(直接返回原始dataframe数据)，则直接返回数据
    #         if (isinstance(corpus_vec, pd.core.frame.DataFrame)):
    #             cluster_text = {0: corpus}
    #             return cluster_text

    #         bert_text = dict(zip(corpus, corpus_vec))

    for index, text in tqdm(enumerate(corpus)):
        max_score=0
        max_cluster=''
        if num_topic == 0:
#             clusters.setdefault(num_topic, []).append(corpus_vec[index])
#             cluster_cores[num_topic] = corpus_vec[index]
            cluster_text.setdefault(num_topic, []).append(text)
            num_topic += 1
        else:
            flag=0
            for key in cluster_text.keys():
                for t in cluster_text[key]:
                    score=cs.cal_similar(text, t) 
                    #print(score,text,t)
                    if score<theta:
                        break
                if t==cluster_text[key][-1] and score>theta:  
                    flag=1
                    cluster_text[key].append(text)
                    break
            if flag==0:# 新簇
                cluster_text.setdefault(num_topic, []).append(text)
                num_topic += 1
                #print('簇个数：%s'% num_topic)
    # cluster_text格式为：{cluster_num1:[text_bag1],cluster_num2:[text_bag2],...}

    return cluster_text
```

```python

import tensorflow as tf

def cosine(q,a):
    pooled_len_1 = tf.sqrt(tf.reduce_sum(q * q, 1))
    pooled_len_2 = tf.sqrt(tf.reduce_sum(a * a, 1))
    pooled_mul_12 = tf.reduce_sum(q * a, 1)
    score = tf.div(pooled_mul_12, pooled_len_1 * pooled_len_2 +1e-8, name="scores")
    return score 
def get_cos_distance(X1, X2):
    # calculate cos distance between two sets
    # more similar more big
    (k, n) = X1.shape
    (m, n) = X2.shape
    # 求模
    X1_norm = tf.sqrt(tf.reduce_sum(tf.square(X1), axis=1))
    X2_norm = tf.sqrt(tf.reduce_sum(tf.square(X2), axis=1))
    # 内积
    X1_X2 = tf.matmul(X1, tf.transpose(X2))
    X1_X2_norm = tf.matmul(tf.reshape(X1_norm, [k, 1]), tf.reshape(X2_norm, [1, m]))
    # 计算余弦距离
    cos = X1_X2 / X1_X2_norm
    return cos
X1=tf.Variable(tf.random_normal(shape=[4,3],mean=0,stddev=1),name='v1')
X2=tf.Variable(tf.random_normal(shape=[4,3],mean=0,stddev=1),name='v1')
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    print(sess.run(X1))
    print(sess.run(X2))
    print(sess.run(cosine(X1, X2)))
    print(sess.run(get_cos_distance(X1, X2)))
```

```python
import numpy as np
a=np.array([1,1,1,1,1,0,1,0,1])
b=np.array([0.95,0.93,0.92,0.99,0.91,0.1,0.87,0.03,0.96])
c=np.array([0.5]*9)
```

```python
def contrastive_loss(y, d):
    tmp = (1-y) * tf.square(d)
    # tmp= tf.mul(y,tf.square(d))
    tmp2 = y * tf.square(tf.maximum((1 - d), 0))
    return tmp+tmp2
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    print(sess.run(tf.reduce_mean(tf.reduce_sum(contrastive_loss(a,c)))))
    print(sess.run(tf.reduce_mean(contrastive_loss(a,b))))
    print(sess.run(tf.reduce_mean(contrastive_loss(a,c))))
```
