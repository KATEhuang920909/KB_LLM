```python
data_primary_q=pd.read_csv('question_all_label.xlsx')
```

```python
data_primary_q
```

```python
primary_question=data_primary_q[data_primary_q.qu_flag==0].question.to_list()
```

```python
primary_question
```

```python
from bert_serving.client import BertClient
from tqdm import tqdm
from sklearn.metrics.pairwise import cosine_similarity
bc=BertClient(ip='10.130.228.46',port=9001,port_out=9002)
q_vec=bc.encode(primary_question)
```

```python
temp_vec=bc.encode(['牙齿包保险是指定哪些诊所'])

max_match_question=['']*5
max_value=[0]*5
for i in tqdm(range(len(q_vec))):
    sim=cosine_similarity(temp_vec,q_vec[i].reshape(1,-1))
    if sim>min(max_value):
        temp_index=max_value.index(min(max_value))
        max_value[temp_index]=sim
        max_match_question[temp_index]=primary_question[i]
```

```python
for k,v in zip(max_value,max_match_question):
    print(k,v)
```
