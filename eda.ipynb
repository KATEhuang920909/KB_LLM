```python
database= open("data/nlpcc-iccpol-2016.kbqa.kb",encoding="utf8").readlines()
```

```python
database=[k.strip() for k in database]
```

```python
database[0]
```

```python
len(database),len(database)//5000000
```

```python
i=0
loop=len(database)//5000000
for i in range(loop):
    with open(f"data/kbqa{i}.txt","w",encoding="utf8") as f:
        for j,unit in enumerate(database[i*5000000:(i+1)*5000000]):
            f.write(str(i*5000000+j)+"\t"+unit)
            f.write("\n")
```

```python
with open(f"data/kbqa{i+1}.txt","w",encoding="utf8") as f:
        for j,unit in enumerate(database[(i+1)*5000000:]):
            f.write(str((i+1)*5000000+j)+"\t"+unit)
            f.write("\n")
```

```python
from tqdm import tqdm_notebook
for unit in tqdm_notebook(database):
    if "中国男子篮球职业联赛" in unit :
        print(unit)
```
